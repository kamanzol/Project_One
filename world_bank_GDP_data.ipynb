```python
 # Dependencies
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

# Adding options to display all of the rows of DataFrame w/o truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
file_path = os.path.join('Resources','GDP_last_50.csv')
gdp_df = pd.read_csv(file_path)
gdp_df = gdp_df.drop(['Series Name', 'Series Code'], axis=1)
gdp_df.head()
```

### All of the data is in string format!

```python
# Everything in the .csv file is in strings
gdp_df.dtypes
```

```python
# split the dfs to convert to numeric
gdp_country_df = gdp_df[['Country Name', 'Country Code']]
gdp_yearly_df = gdp_df[list(gdp_df.columns)[2:]]

# Replace '..' with NaN and convert to numeric
gdp_yearly_df = gdp_yearly_df.replace('..',np.nan)
gdp_yearly_df = gdp_yearly_df.apply(pd.to_numeric)

# Convert to US$ in Billions
gdp_yearly_df = gdp_yearly_df/10E9

# Re-combine the country and gdp data
gdp_clean_df = pd.concat([gdp_country_df, gdp_yearly_df], axis=1)

gdp_clean_df.head()
```

```python
# Rename YEAR columns
# Create a dictionary to update year titles in the column headers
a_dict={}
year_headers = (list(gdp_clean_df)[2:])
for year in year_headers:
    a_dict.update({year:year[0:4]})
gdp_clean_df = gdp_clean_df.rename(columns=a_dict)
gdp_clean_df.head()
```

```python
gdp_clean_df.tail(10)
```

```python
# drop last 5 all NaN rows becuase there is no data in there
gdp_clean_df = gdp_clean_df.dropna(how = 'all')
gdp_clean_df.tail()
```

```python
# Bring in latitudes and longitudes for the countries
from config import g_key
import requests
import json

country_Lat_Lng_df = pd.DataFrame({'Country Name':[],\
                               'Latitude':[],\
                               'Longitude':[]})
# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

request_counter = 0
for a_country in list(gdp_clean_df['Country Name']):
    params = {"address": a_country, "key": g_key}
    response = requests.get(base_url, params=params)
    response = response.json()
    
#     print(a_country)
    if response:
        try:
            latitude = (response['results'][0]['geometry']['location']['lat'])
            longitude = (response['results'][0]['geometry']['location']['lat'])

            response_items_df = pd.DataFrame([[a_country, \
                                              latitude, \
                                              longitude]], columns=list(country_Lat_Lng_df.columns))

            country_Lat_Lng_df = country_Lat_Lng_df.append(response_items_df)
        except:
            continue
    else:
        print(f'{a_country} not found. Skipping')
        continue
    
    request_counter = request_counter + 1
#     if (request_counter == 10):
#         break   
```

```python
country_Lat_Lng_df.head()
```

```python
# Merge the two dfs and write out as .csv file
gdp_clean_df = pd.merge(gdp_clean_df,country_Lat_Lng_df, on='Country Name')
gdp_clean_df.head()
```
